## Train a model on pre-2010 data and evaluate its performance on the 2010 houses.

- Create the training and testing data
- Train data is pre-2010 data, test data is 2010 houses.

```python
new_df1 = df1[fixed_predictors]
```

```python
new_df1.shape
```

```python
new_df1
```

```python
new_df1_dummy = pd.get_dummies(new_df1)
```

```python
new_df1_dummy.shape
```

```python
# specify predictor matrix and target variable
X = new_df1_dummy
y = df1.SalePrice_log.values
```

```python
new_df1_dummy.YrSold.unique()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# standardize the predictor
scaler = StandardScaler()
Xs = scaler.fit_transform(X)
```

```python
X_train = Xs[new_df1_dummy.YrSold.isin([2008, 2007, 2006, 2009])]
X_test = Xs[new_df1_dummy.YrSold == 2010]

y_train = y[new_df1_dummy.YrSold.isin([2008, 2007, 2006, 2009])]
y_test = y[new_df1_dummy.YrSold == 2010]
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

# Fit Regression Models

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, RidgeCV, LassoCV, ElasticNetCV
```

```python
# Ridge Regression
ridge_alphas = np.logspace(0, 5, 200)

optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(X_train, y_train)

print(optimal_ridge.alpha_)
```

```python
# Cross-validate the Ridge regression R^2 with the optimal alpha
ridge = Ridge(alpha=optimal_ridge.alpha_)

ridge_scores = cross_val_score(ridge, X_train, y_train, cv=10)

print(ridge_scores)
print(np.mean(ridge_scores))
```

```python
# Lasso Regression
```

```python
# Find an optimal value for Lasso regression alpha using LassoCV
optimal_lasso = LassoCV(n_alphas=500, cv=10, verbose=1, n_jobs=2)
optimal_lasso.fit(X_train, y_train)

print(optimal_lasso.alpha_)
```

```python
# Cross-validate the Lasso R^2 with the optimal alpha
lasso = Lasso(alpha=optimal_lasso.alpha_)

lasso_scores = cross_val_score(lasso, X_train, y_train, cv=10)

print(lasso_scores)
print(np.mean(lasso_scores))
```

The lasso performs slightly better than the Ridge

```python
# Ridge Regression
# Find an optimal value for Ridge regression alpha using RidgeCV
ridge_alphas_2 = np.logspace(0, 5, 200)

optimal_ridge_2 = RidgeCV(alphas=ridge_alphas_2, cv=10)
optimal_ridge_2.fit(X2_train, y2_train)

print(optimal_ridge_2.alpha_)
```

```python
# Cross-validate the Ridge regression R^2 with the optimal alpha
ridge_2 = Ridge(alpha=optimal_ridge_2.alpha_)

ridge_scores_2 = cross_val_score(ridge_2, X2_train, y2_train, cv=10)

print(ridge_scores_2)
print(np.mean(ridge_scores_2))
```

```python
# Lasso Regression
# Find an optimal value for Lasso regression alpha using LassoCV

optimal_lasso_2 = LassoCV(n_alphas=300, cv=10, verbose=1)
optimal_lasso_2.fit(X2_train, y2_train)

print(optimal_lasso_2.alpha_)
```
